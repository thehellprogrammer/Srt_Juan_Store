```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
tienda.info()


```

```python
ventas1 = tienda['Precio'].sum()
print(f"Las ventas de la tienda 1 es:{ventas1}")
ventas2 = tienda2['Precio'].sum()
print(f"Las ventas de la tienda 2 es:{ventas2}")
ventas3 = tienda3['Precio'].sum()
print(f"Las ventas de la tienda 3 es:{ventas3}")
ventas4 = tienda4['Precio'].sum()
print(f"Las ventas de la tienda 4 es:{ventas4}")

envio1 = round (tienda ['Costo de envío'].sum() / len(tienda ['Costo de envío']),2)
print(f"El costo de envio promedio de la tienda 1 es:{envio1}")
envio2 = round (tienda2 ['Costo de envío'].sum() / len(tienda2 ['Costo de envío']),2)
print(f"El costo de envio promedio de la tienda 2 es:{envio2}")
envio3 = round (tienda3 ['Costo de envío'].sum() / len(tienda3 ['Costo de envío']),2)
print(f"El costo de envio promedio de la tienda 3 es:{envio3}")
envio4 = round (tienda4 ['Costo de envío'].sum() / len(tienda4 ['Costo de envío']),2)
print(f"El costo de envio promedio de la tienda 4 es:{envio4}")

Reseña1 = round (tienda ['Calificación'].sum() / len(tienda ['Calificación']),2)
print(f"La reseña promedio de la tienda 1 es:{Reseña1}")
Reseña2 = round (tienda2 ['Calificación'].sum() / len(tienda2 ['Calificación']),2)
print(f"La reseña promedio de la tienda 2 es:{Reseña2}")
Reseña3 = round (tienda3 ['Calificación'].sum() / len(tienda3 ['Calificación']),2)
print(f"La reseña promedio de la tienda 3 es:{Reseña3}")
Reseña4 = round (tienda4 ['Calificación'].sum() / len(tienda4 ['Calificación']),2)
print(f"La reseña promedio de la tienda 4 es:{Reseña4}")



```
